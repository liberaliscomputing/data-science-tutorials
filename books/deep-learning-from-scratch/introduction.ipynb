## 1.2 Drawing Graphs

```python
# Prepare data
x = np.arange(0, 6, 0.1)
sin_x = np.sin(x)
cos_x = np.cos(x)
```

```python
# Draw graphs in one pane
plt.plot(x, sin_x)
plt.plot(x, cos_x, linestyle='--', c='r')
plt.legend(['Sin', 'Cos'])
plt.show()
```

```python
# Draw graphs in multiple panes
plt.subplot(2, 1, 1)
plt.plot(x, sin_x)
plt.legend(['Sin'])

plt.subplot(2, 1, 2)
plt.plot(x, cos_x, linestyle='--', c='r')
plt.legend(['Cos'])

plt.show()
```
